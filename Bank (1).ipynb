```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(bank['Balance'], bins=30,  color = 'green',edgecolor='black')  # bins can be adjusted
plt.title('Distribution of Account Balances')          # Histogram title
plt.xlabel('Balance')                                   # X-axis label
plt.ylabel('Frequency')                                 # Y-axis label
# plt.grid(axis='y', alpha=0.3)                          # Optional: adds horizontal grid
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(bank["Balance"])
plt.title("Balance Distribution with IQR Outlier Detection")
plt.ylabel("Balance")
plt.show()
```

```python

Q1 = bank["Balance"].quantile(0.25)
Q3 = bank["Balance"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = bank[(bank["Balance"] < lower) |
                (bank["Balance"] > upper)]
print(Q1)
print(Q3)
print(IQR)
print(lower)
print(upper)
print(outliers)
outliers["Balance"]
```

```python
bank['Balance'].max()
```

```python
len(outliers)
```

```python
mask = (bank["Balance"] < lower) | (bank["Balance"] > upper)
print(mask.value_counts())
print("Lower:", lower)
print("Upper:", upper)
```
